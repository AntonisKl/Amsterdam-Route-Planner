```python
import folium
import pyproj
import requests
from openrouteservice import client
from shapely import geometry
from shapely.geometry import Point, LineString, Polygon, MultiPolygon
import geojson
import geopandas as gpd
```

```python
api_key = '5b3ce3597851110001cf6248aec4a63d15b347b0a8da0d9a7725d391'
ors = client.Client(key=api_key)
```

```python
request_params = {'coordinates': [[4.899431, 52.379189],
                                  [4.8852, 52.3600]],
                  'format_out': 'geojson',
                  'profile': 'foot-walking',
                  'preference': 'shortest',
                  'instructions': 'true', }

route_normal = ors.directions(**request_params)
```

```python
m = folium.Map(
    location=[52.3676, 4.9041],
    tiles="cartodbpositron",
    zoom_start=13
)
```

```python
def style_function(color):
    return lambda feature: dict(color=color,
                                weight=3,
                                opacity=0.5)


folium.features.GeoJson(data=route_normal,
                        name='Route without construction sites',
                        style_function=style_function('#FF0000'),
                        overlay=True).add_to(m)
m
```

```python
# Buffer route with 0.009 degrees (really, just too lazy to project again...)
route_buffer = LineString(route_normal['features'][0]['geometry']['coordinates']).buffer(0.00015)
folium.features.GeoJson(data=geometry.mapping(route_buffer),
                        name='Route Buffer',
                        style_function=style_function('#FFFF00'),
                        overlay=True).add_to(m)
m
```

```python
accessibility = requests.get('https://sidewalk-amsterdam.cs.washington.edu/v2/access/score/streets?lng1=4.769276&lat1=52.300761&lng2=5.004452&lat2=52.426341').content

accessibility_geojson = geojson.loads(accessibility)
accessibility_geojson
```

```python
accessibility_gdf = gpd.GeoDataFrame.from_features(accessibility_geojson['features'])
accessibility_gdf
```

```python
def calculate_accessibility_level(score):
    if score < 0.25:
        return 0
    elif score < 0.5:
        return 1
    elif score < 0.75:
        return 2
    else:
        return 3


accessibility_gdf['score_0to3'] = accessibility_gdf['score'].apply(calculate_accessibility_level)
accessibility_gdf
```

```python
bad_accessibility_streets = accessibility_gdf[accessibility_gdf['score_0to3'] >= 2]['geometry']
bad_accessibility_streets.crs = 'urn:ogc:def:crs:OGC:1.3:CRS84'

bad_accessibility_streets
```

```python
streets_to_avoid = []
for bad_accessibility_street in bad_accessibility_streets:
    # poly = Polygon(site_poly)
    if route_buffer.intersects(bad_accessibility_street):
        folium.features.Marker(list(reversed(bad_accessibility_street.centroid.coords[0]))).add_to(m)
        folium.GeoJson(bad_accessibility_street.buffer(0.00015), name="accessibility",
               style_function=lambda feature: {'color': 'red'}).add_to(m)
        streets_to_avoid.append(bad_accessibility_street.buffer(0.0002))
m
```

```python
request_params['options'] = {'avoid_polygons': geometry.mapping(MultiPolygon(streets_to_avoid))}
route_detour = ors.directions(**request_params)

folium.features.GeoJson(data=route_detour,
                        name='Route with construction sites',
                        style_function=style_function('#00FF00'),
                        overlay=True).add_to(m)

m.add_child(folium.map.LayerControl())
m
```
